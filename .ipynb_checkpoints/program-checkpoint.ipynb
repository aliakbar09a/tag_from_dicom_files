```python
import pydicom
```

```python
file1 = pydicom.dcmread('ttfm.dcm')
file2 = pydicom.dcmread('bmode.dcm')
```

```python
ttfm = file1.dir()
bmode = file2.dir()
```

```python
ttfm_tag = [file1.data_element(tag_name).tag for tag_name in ttfm]
bmode_tag = [file1.data_element(tag_name).tag for tag_name in bmode]
```

```python
with open('ttfm.txt', 'w') as f:
    f.writelines('tag\t\t' + 'tag name' + '\t\t\t\tvalue\n')
    for tag in ttfm_tag:
        f.writelines(str(file1[tag]))
        f.write('\n')
with open('bmode.txt', 'w') as f:
    f.writelines('tag\t\t' + 'tag name' + '\t\t\t\tvalue\n')
    for tag in bmode_tag:
        f.writelines(str(file2[tag]))
        f.write('\n')
```
